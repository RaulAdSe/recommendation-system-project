# Movie Recommendation System Demo

This notebook tests, compares, and visualizes the results of different recommendation algorithms implemented in the `recommendation.py` script. 

## Objectives
1. Test the performance of various recommendation algorithms.
2. Compare their effectiveness using evaluation metrics (e.g., RMSE).
3. Generate and visualize recommendations for sample users.


```python
# Load the ratings and movies datasets
ratings = load_data('../data/ratings.csv')
movies = load_data('../data/movies.csv')

# Display basic information about the datasets
print(f"Number of ratings: {ratings.shape[0]}")
print(f"Number of movies: {movies.shape[0]}")
print(f"Number of unique users: {ratings['userId'].nunique()}")
print(f"Number of unique movies rated: {ratings['movieId'].nunique()}")
```

```python
# Create the user-item matrix for collaborative filtering
user_item_matrix = create_user_item_matrix(ratings)

print("User-Item Matrix (Sample):")
display(user_item_matrix.head())
```

```python
# Compute user similarity matrix
user_similarity = compute_user_similarity(user_item_matrix)

print("User Similarity Matrix (Sample):")
display(user_similarity.head())

# Compute item similarity matrix using TF-IDF for genres
tf_idf_matrix = compute_tf_idf(movies)
item_similarity = tf_idf_matrix.dot(tf_idf_matrix.T)

print("Item Similarity Matrix (Sample):")
display(item_similarity.head())
```

```python
# Import cosine_similarity from scikit-learn
from sklearn.metrics.pairwise import cosine_similarity


print("Dimensions of user_item_matrix_filled:", user_similarity.shape)

print("Dimensions of item_similarity:", item_similarity.shape)

# Compute mean rating for each user
print("Dimensions of mean_user_rating:", mean_user_rating.shape)

# Compute the rating difference from the mean
ratings_diff = (user_item_matrix_filled.T - mean_user_rating).T
print("Dimensions of ratings_diff:", ratings_diff.shape)

# Debugging the alignment issue by printing shapes before dot product
print("Dimensions for dot product operation:")
print("Ratings difference matrix shape:", ratings_diff.shape)
print("Item similarity matrix shape:", item_similarity.shape)
```

```python
# Compute user similarity matrix using cosine similarity
user_similarity = pd.DataFrame(cosine_similarity(user_item_matrix_filled.fillna(0)), 
                               index=user_item_matrix.index, 
                               columns=user_item_matrix.index)

# Function to predict ratings based on user similarity
def predict_ratings_user_based(user_item_matrix, user_similarity):
    """Predict ratings based on user similarity."""
    # Fill NaN values with 0 for calculations
    user_item_matrix_filled = user_item_matrix.fillna(0)
    
    # Compute mean rating for each user
    mean_user_rating = user_item_matrix_filled.mean(axis=1)
    
    # Compute the rating difference from the mean
    ratings_diff = (user_item_matrix_filled.T - mean_user_rating).T
    
    # Predict ratings using user similarity
    pred = mean_user_rating.values[:, np.newaxis] + user_similarity.dot(ratings_diff.fillna(0)) / np.array([np.abs(user_similarity).sum(axis=1)]).T
    
    # Convert to DataFrame with appropriate index and column names
    pred_df = pd.DataFrame(pred, index=user_item_matrix.index, columns=user_item_matrix.columns)
    
    return pred_df

# Predict ratings using user-based collaborative filtering
predicted_ratings_user_cf = predict_ratings_user_based(user_item_matrix, user_similarity)
print("Predicted Ratings (User-Based CF) (Sample):")
display(predicted_ratings_user_cf.head())

# Predict ratings using item-based collaborative filtering
predicted_ratings_item_cf = predict_ratings_user_based(user_item_matrix.T, item_similarity)
print("Predicted Ratings (Item-Based CF) (Sample):")
display(predicted_ratings_item_cf.head())
```

```python
# Predict ratings using SVD
predicted_ratings_user_svd = svd_recommendation(user_item_matrix, k=50)
print("Predicted Ratings (User-Based SVD) (Sample):")
display(predicted_ratings_user_svd.head())
```

```python
# Function to predict ratings based on item similarity
def predict_ratings_item_based(user_item_matrix, item_similarity):
    """Predict ratings based on item similarity."""
    # Transpose the user-item matrix to align items with rows
    user_item_matrix_filled = user_item_matrix.fillna(0)  # Fill NaNs with 0 for calculation

    # Initialize the predicted ratings matrix
    pred = np.zeros(user_item_matrix.shape)

    # Loop over all users
    for i in range(user_item_matrix_filled.shape[0]):
        # Loop over all items
        for j in range(user_item_matrix_filled.shape[1]):
            # Only predict if the user has not rated the item (i.e., rating is NaN in the original matrix)
            if user_item_matrix.iloc[i, j] == 0:
                # Get the similarities between the current item and all other items
                sim_scores = item_similarity[j]
                # Get the ratings the user has given to other items
                user_ratings = user_item_matrix_filled.iloc[i]
                # Compute the weighted sum of the similarities
                weighted_sum = np.dot(sim_scores, user_ratings)
                # Compute the sum of the absolute similarities
                sum_of_similarities = np.abs(sim_scores).sum()

                # Predict the rating for the current item
                if sum_of_similarities != 0:
                    pred[i, j] = weighted_sum / sum_of_similarities
                else:
                    pred[i, j] = 0  # If no similar items, predict 0

    # Convert to DataFrame with appropriate index and column names
    pred_df = pd.DataFrame(pred, index=user_item_matrix.index, columns=user_item_matrix.columns)
    
    return pred_df

# Use the function to predict item-based ratings
predicted_ratings_item_cf = predict_ratings_item_based(user_item_matrix, item_similarity)
print("Predicted Ratings (Item-Based CF) (Sample):")
display(predicted_ratings_item_cf.head())
```

```python
# Convert to DataFrame with appropriate index and column names
predicted_ratings_item_cf = pd.DataFrame(predicted_ratings_item_cf, index=user_item_matrix.index, columns=user_item_matrix.columns)

# Display the predicted ratings (item-based CF)
print("Predicted Ratings (Item-Based CF) (Sample):")
display(predicted_ratings_item_cf.head())
```

```python
# Generate item-based recommendations for a sample user (User 1)
sample_user_id = 1
top_n_item_based_recommendations = recommend_items(sample_user_id, user_item_matrix, item_similarity, num_recommendations=5)

print(f"Top 5 Item-Based Recommendations for User {sample_user_id}:")
display(top_n_item_based_recommendations)
```

```python
# Evaluate collaborative filtering model
rmse_cf = evaluate_model(user_item_matrix, predicted_ratings_cf)
print(f'Collaborative Filtering RMSE: {rmse_cf:.4f}')

# Evaluate SVD-based model
rmse_svd = evaluate_model(user_item_matrix, predicted_ratings_svd)
print(f'SVD-Based RMSE: {rmse_svd:.4f}')

# Plot RMSE comparison for different models
models = ['User-Based CF', 'SVD']
rmse_values = [rmse_cf, rmse_svd]

plt.figure(figsize=(8, 6))
plt.bar(models, rmse_values, color=['skyblue', 'green'])
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.title('Model Performance Comparison (RMSE)')
plt.show()
```

```python
# Function to generate top-N recommendations
def recommend_top_n(user_id, user_item_matrix, predicted_ratings, num_recommendations=5):
    user_ratings = predicted_ratings.loc[user_id].sort_values(ascending=False).head(num_recommendations)
    movie_titles = movies.set_index('movieId').loc[user_ratings.index]['title']
    recommendations = pd.DataFrame({'Movie': movie_titles, 'Predicted Rating': user_ratings.values})
    return recommendations

# Generate recommendations for a sample user (User 1)
sample_user_id = 1
top_n_recommendations_cf = recommend_top_n(sample_user_id, user_item_matrix, predicted_ratings_cf)
top_n_recommendations_svd = recommend_top_n(sample_user_id, user_item_matrix, predicted_ratings_svd)

print(f"Top {len(top_n_recommendations_cf)} Recommendations for User {sample_user_id} (Collaborative Filtering):")
display(top_n_recommendations_cf)

print(f"Top {len(top_n_recommendations_svd)} Recommendations for User {sample_user_id} (SVD):")
display(top_n_recommendations_svd)
```
