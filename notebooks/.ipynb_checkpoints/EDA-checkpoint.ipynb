```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for seaborn
sns.set(style='whitegrid')
```

```python
# Load the ratings dataset
ratings = pd.read_csv('../data/ratings.csv')
print("Ratings Data (First 5 Rows):")
display(ratings.head())

# Load the movies dataset
movies = pd.read_csv('../data/movies.csv')
print("\nMovies Data (First 5 Rows):")
display(movies.head())
```

```python
# Shape of the datasets
print(f"Shape of ratings dataset: {ratings.shape}")
print(f"Shape of movies dataset: {movies.shape}")

# Column names and data types
print("\nRatings Dataset Info:")
ratings.info()

print("\nMovies Dataset Info:")
movies.info()

# Checking for missing values
print("\nMissing Values in Ratings Dataset:")
print(ratings.isnull().sum())

print("\nMissing Values in Movies Dataset:")
print(movies.isnull().sum())
```

```python
# Descriptive statistics for ratings
print("Descriptive Statistics for Ratings:")
display(ratings.describe())

# Descriptive statistics for movie genres
print("\nCount of Unique Genres:")
unique_genres = pd.Series([genre for sublist in movies['genres'].fillna('').apply(lambda x: x.split('|')).tolist() for genre in sublist])
print(unique_genres.value_counts())
```

```python
# Distribution of ratings
plt.figure(figsize=(8, 6))
sns.countplot(x='rating', data=ratings)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Users tend to rate movies positively, with a significant spike at 4.0, indicating that most ratings are relatively high. Very few users give the lowest or highest possible ratings (0.5 or 5.0), suggesting a bias towards moderate and slightly high ratings.

```python
# Number of ratings per user
ratings_per_user = ratings.groupby('userId').size()
plt.figure(figsize=(10, 6))
sns.histplot(ratings_per_user, bins=30, kde=False, color='purple')
plt.title('Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

A large portion of users have rated only a few movies, while very few users have rated many movies. This indicates a high level of sparsity in the user-item matrix, which is typical in recommendation system datasets. The dataset may be susceptible to the "cold start" problem, where new or inactive users lack enough data for accurate predictions.

```python
# Number of ratings per movie
ratings_per_movie = ratings.groupby('movieId').size()
plt.figure(figsize=(10, 6))
sns.histplot(ratings_per_movie, bins=30, kde=False, color='teal')
plt.title('Number of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Movies')
plt.show()
```

A small number of movies receive a large number of ratings, while the vast majority of movies have very few ratings. This suggests a "long tail" distribution, where a few popular movies dominate user interactions, and many niche or less popular movies are rated by only a few users.

```python
# Top 10 most rated movies
top_rated_movies = ratings.groupby('movieId').size().sort_values(ascending=False).head(10)
top_rated_movies = top_rated_movies.reset_index().merge(movies[['movieId', 'title']], on='movieId')
print("Top 10 Most Rated Movies:")
display(top_rated_movies)
```

```python
# Average rating per movie
avg_rating_per_movie = ratings.groupby('movieId')['rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.histplot(avg_rating_per_movie, bins=30, kde=True, color='coral')
plt.title('Average Rating per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Number of Movies')
plt.show()
```

There is a tendency for movies to receive average ratings that cluster around 3.0 and 4.0, with fewer movies rated extremely low or high. This suggests that while most movies are generally rated positively, there are distinct segments where movies receive varied ratings, indicating diversity in user preferences.

```python
# Correlation between the number of ratings and average rating
ratings_summary = ratings.groupby('movieId').agg({'rating': ['mean', 'count']})
ratings_summary.columns = ['avg_rating', 'num_ratings']

# Scatter plot to visualize correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='num_ratings', y='avg_rating', data=ratings_summary, alpha=0.6)
plt.title('Correlation between Number of Ratings and Average Rating')
plt.xlabel('Number of Ratings')
plt.ylabel('Average Rating')
plt.show()

# Print correlation coefficient
correlation = ratings_summary['num_ratings'].corr(ratings_summary['avg_rating'])
print(f"Correlation between number of ratings and average rating: {correlation:.2f}")
```

There is a weak correlation between the number of ratings a movie receives and its average rating. This suggests that the popularity of a movie (number of ratings) does not strongly influence its overall rating. Popular movies are not necessarily highly rated, and vice versa. However, movies with very few ratings tend to show more variability in their average ratings, likely due to less aggregated feedback.

The dataset exhibits common characteristics seen in recommendation systems: rating distributions skewed towards higher values, sparsity in user and movie interactions, and a "long tail" of movies with very few ratings.
